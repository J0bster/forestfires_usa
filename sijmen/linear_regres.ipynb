```python
import pandas as pd
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from scipy.stats import linregress
import os
```

```python
mean_df = pd.read_csv("meanCO2_values.csv")
mean_df["m_y"] = mean_df["year"].astype(str) +"-" + mean_df["month"].astype(str)
plt.plot(mean_df["m_y"], mean_df["co2_correct"])

slope, intercept, _, _, _ = linregress(mean_df["m_y"].index, mean_df["co2_correct"]) # getting linear regresion data of uncleaned datafile

x_vals =[]
y_vals =[]

for i in range(len(mean_df["m_y"])):
    x_vals.append(mean_df["m_y"].index)
    y_vals.append(mean_df["m_y"].index * slope + intercept)

#plt.plot(mean_df["m_y"], mean_df["co2_correct"])
plt.plot(x_vals, y_vals)
plt.show()
```
